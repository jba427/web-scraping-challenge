```python
#Dependencies
import pymongo
from bs4 import BeautifulSoup as bs
import pandas as pd
from splinter import Browser
from flask import Flask, render_template
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

return_list = {}
```

```python
# scrape nasa.gov
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
find1 = soup.find("div", class_="list_text")

print(find1)
```

```python
#news_title = find1.find("div", class_="content_title").text
return_list["news_title"] = find1.find("div", class_="content_title").text
print(return_list["news_title"])
```

```python
return_list["news_p"] = find1.find("div", class_="article_teaser_body").text
print(return_list["news_p"])
```

```python
# scrape jpl.nasa.gov
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
find1 = soup.find("body", class_="dark_background")

print(find1)
```

```python
mars_source = find1.find("div", class_="image_and_description_container"). \
    find("img", class_="thumb")
print(mars_source)
```

```python
return_list["featured_image_url"] = url + mars_source['src']
print(return_list["featured_image_url"])
```

```python
# scrape twitter for mars weather
url = 'https://twitter.com/marswxreport?lang=en'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
find1 = soup.findAll("section", class_="css-1dbjc4n")
print(find1)
```

```python
# scrape https://space-facts.com/mars/
url = 'https://space-facts.com/mars/'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
mars_table = pd.read_html(url)
mars_table
```

```python
df = mars_table[0]
df
```

```python
html_table = df.to_html()
html_table
```

```python
return_list["html_table"] = html_table.replace('\n','')

print(return_list["html_table"])
```

```python
# scrape https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
# variables to return
# news_title
# news_p
# featured_image_url
# <missing piece from twitter> mars_weather
# html_table
# <missing data from astrogeology> hemisphere_image_urls
return_list
```

```python
#Setup mongo connection
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.nasa_db
collection = db.nasa
```

```python
#insert into Mongo
collection.insert_one(return_list)
```

```python
#Display item from Mongo
#moon = db.nasa.find()
mars = list(collection.find())

print(mars)
```

```python
mars_df = pd.DataFrame(mars)
```
